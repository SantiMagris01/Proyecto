```python
import pandas as pd
```

```python
df = pd.read_csv("Dataset\movies_dataset.csv")
```

```python
df
```

```python
#Convertimos los campos "revenue" y "budget" en tipo numerico
df["revenue"] = pd.to_numeric(df["revenue"], errors="coerce")
df["budget"] = pd.to_numeric(df["budget"], errors="coerce")
```

```python
#Rellenar campos "Revenue" y "budget"
df["budget"] = df["budget"].fillna(0)
df["revenue"] = df["revenue"].fillna(0)
```

```python
#Convertimos las fechas al fromato (AAAA-mm-dd)
df["release_date"] = pd.to_datetime(df["release_date"], format= "%Y-%m-%d", errors="coerce")
```

```python
#Eliminamos los valores nulos de "realse date"
df = df.dropna(subset=["release_date"])
```

```python
#Creamos la columna "release_year"
df["release_year"]= df["release_date"].dt.year
```

```python
#Creamos la columna "return"
df["return"] = df["revenue"].fillna(0) / df["budget"].fillna(0)
```

```python
df["return"].fillna(0, inplace=True)
```

```python
#Eliminamos las columnas que no van a ser utilizadas
columns_to_drop = ["video", "imdb_id", "adult", "original_title", "poster_path", "homepage"]
df = df.drop(columns=columns_to_drop)
```

```python
import re
```

```python
#Extraemos los diccionarios y los convertimos en columnas independientes
pattern = r"{'id': (\d+), 'name': '([^']+)', 'poster_path': '([^']+)', 'backdrop_path': '([^']+)'"
extracted_data = df["belongs_to_collection"].str.extract(pattern)
```

```python
#Renombramos las columnas
extracted_data.columns = ["collection_id", "collection_name", "collection_poste_path", "collection_backdrop_path"]
```

```python
extracted_data
```

```python
#Concatenamos los datos extraidos con el dataset original
df= pd.concat([df, extracted_data], axis=1)
```

```python
#Eliminamos la columna "belongs_to_collection"
df.drop(columns="belongs_to_collection", inplace=True)
```

```python
#Eliminamos columnas que no nos interesan
df.drop(columns=["collection_id", "collection_poste_path", "collection_backdrop_path"], inplace=True)
```

```python
import ast
```

```python
#Convertimos las cadenas de texto en listas de diccionarios
df["genres"] = df["genres"].apply(ast.literal_eval)
```

```python
#Definimos una funcion que extrae los elementos de la clave 'name'
def extract_names(list):
    names = [x["name"] for x in list]
    return names
```

```python
#Aplicamos la funcion
df["genres_names"] = df["genres"].apply(extract_names)
```

```python
#Eliminamos la columna "genres"
df.drop(columns="genres", inplace=True)
```

```python
df
```

```python
#Repetimos el proceso con las demas columnas
df["production_companies"] = df["production_companies"].apply(ast.literal_eval)
df["production_companies_names"] = df["production_companies"].apply(extract_names)
df.drop(columns="production_companies", inplace=True)
```

```python
df["production_countries"] = df["production_countries"].apply(ast.literal_eval)
df["production_countries_names"] = df["production_countries"].apply(extract_names)
df.drop(columns="production_countries", inplace=True)
```

```python
df["spoken_languages"] = df["spoken_languages"].apply(ast.literal_eval)
df["spoken_languages_names"] = df["spoken_languages"].apply(extract_names)
df.drop(columns="spoken_languages", inplace=True)
```

```python
#Desanidamos las columnas 'cast' y 'crew'
#Creamos otro df donde vamos a guardar los datos desanidados
df2_modificado = df2
```

```python
df2_modificado["cast"] = df2_modificado["cast"].apply(ast.literal_eval)
df2_modificado["cast_names"] = df2_modificado["cast"].apply(extract_names)
df2_modificado.drop(columns="cast", inplace=True)
```

```python

df2_modificado["crew"] = df2_modificado["crew"].apply(ast.literal_eval)
# Definimos una función para extraer el nombre del director
def extract_director_name(crew):
    for member in crew:
        if member['job'] == 'Director':
            return member['name']
    return None

# Aplicamos la función a la columna 'crew' y creamos una nueva columna 'director_name'
df2_modificado['director_name'] = df2_modificado['crew'].apply(extract_director_name)

```

```python
df2_modificado.drop(columns='crew',inplace=True)
```

```python
df2_modificado.drop(columns='id',inplace=True)
```
